III. **MODELO DE DATOS**

Estructura del Data Warehouse:
Después de explorar el dataset, decidimos que la mejor forma de proceder es con la implementación de un Modelo Dimensional de Estrella, basado en 6 tablas de dimensión y 1 tabla de hechos. Las razones que justifican la elección del Modelo Dimensional son:
1.	Estructura simplificada: El esquema en estrella tiene una estructura simple y fácil de entender, lo que lo hace fácil de manejar y mantener.
2.	Rendimiento optimizado: Con la estructura logramos optimización del rendimiento al reducir la complejidad de las consultas y mejorar los tiempos de respuesta. 
3.	Acceso rápido a los datos: El esquema en estrella permite un acceso rápido a los datos, ya que se pueden realizar consultas complejas de datos utilizando un conjunto limitado de tablas de hechos y dimensiones.
4.	Fácil de implementar: Es de fácil de implementación debido a que su estructura es simple, por lo tanto, es una buena opción para empresas que desean una solución de almacenamiento fácil de implementar y de mantener.
5.	Escalabilidad: Esto lo hace ideal para empresas que tienen un gran volumen de datos y necesitan una solución que pueda manejar el crecimiento de los datos con el tiempo.
6.	Facilidad de uso: El esquema en estrella es fácil de usar, ya que se puede acceder a los datos utilizando consultas SQL simples y herramientas de inteligencia empresarial populares.

```python
try:
    response = aws_conn.create_db_instance(
            AllocatedStorage=10,
            DBName=config.get('RDS_MYSQL', 'DB_NAME'),
            DBInstanceIdentifier=rdsIdentifier,
            DBInstanceClass="db.t3.micro",
            Engine="mysql",
            MasterUsername=config.get('RDS_MYSQL', 'DB_USER'),
            MasterUserPassword=config.get('RDS_MYSQL', 'DB_PASSWORD'),
            Port=int(config.get('RDS_MYSQL', 'DB_PORT')),
            VpcSecurityGroupIds=[config.get('VPC', 'SECURITY_GROUP')],
            PubliclyAccessible=True
        )
    print(response)
except aws_conn.exceptions.DBInstanceAlreadyExistsFault as ex:
    print("La Instancia de Base de Datos ya Existe.")
```

```python
try:
     instances = aws_conn.describe_db_instances(DBInstanceIdentifier=rdsIdentifier)
     RDS_DW_HOST = instances.get('DBInstances')[0].get('Endpoint').get('Address')
     print(RDS_DW_HOST)
except Exception as ex:
     print("La instancia de base de datos no existe o aun no se ha terminado de crear.")
     print(ex)
```

```python
import CreacionDWquery
import mysql.connector as mysqlC
try:
    myDw = mysqlC.connect(
    host=RDS_DW_HOST, 
    user=config.get('RDS_MYSQL', 'DB_USER'),
    password=config.get('RDS_MYSQL', 'DB_PASSWORD'),
    database=config.get('RDS_MYSQL', 'DB_NAME')
    )

    mycursor = myDw.cursor()
    mycursor.execute(CreacionDWquery.CREATE_DW, multi=True)
    myDw.commit()
    print("Data Warehouse Creado Exitosamente")
except Exception as ex:
    print("ERROR: Error al crear la base de datos.")
    print(ex)

```
